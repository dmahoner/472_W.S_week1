Problem 7

```python
# Function to calculate daily water intake
def calculate_water_intake(weight, activity_level):
  # Base water intake  (in liters ) calculation based on weight
  base_intake = weight * 0.033

  # Adjusting water intake based on activity level
  if activity_level == 'low':
    return base_intake
  elif activity_level == 'moderate':
    return base_intake + 0.5
  elif activity_level == 'high':
    return base_intake + 1.0
  else:
    return None  # Invalid activity level

# Main interactive program
print("Welcome to the daily Water Intake Calculator!")
while True:
  try:
    # Get user weight
    weight = float(input("Enter your weight in kg: "))
    break
  except ValueError:
    print("Please enter a valid number for weight.")

while True:
  # Get user activity level
  activity_level = input("Enter your activity level (low, moderate, high): ").lower()

  # Calculate water intake
  daily_water = calculate_water_intake(weight, activity_level)
  if daily_water:
    print(f"Based on your weight and activity level, you should drink approximately {daily_water:.2f} liters of water per day.")
    break  # Exit the loop after a valid input
  else:
    print("Please enter a valid activity level (low, moderate, high).")

print("Stay hydrated!")
```
